# prepare our data for model --

```python
# he CarName column in the dataset represent the company of the car followed by the model name of the car. 
# We are interested only in the company of that car. Hence we the to transform the data values of that column.
```

```python
carcompany = []
for model in df['CarName']:
    carcompany.append(str(model.split(' ')[0]))

df['carcompany'] = np.array(carcompany) # Creating a new column for the company name 
```

```python
with plt.style.context(('ggplot')):
    plt.figure(figsize=(15,7))
    plt.plot(df.groupby(['carcompany']).median()["price"][:10])
    plt.show()
```

```python
df.CarName=df.CarName.astype('category').cat.codes
df.head()  # convert the CarName columns into numeric form 
# so that it unique.. 
```

```python
df.carcompany = df.carcompany.astype('category').cat.codes
df.head()
```

```python
df['CarName'].value_counts()
```

```python
with plt.style.context(('ggplot')):
    plt.figure(figsize=(10,7))
    plt.title("Types of Cars")
    sns.countplot(data=df,x=df['carbody'])
    plt.show()
```

```python
# convert all the coumns data into category format so that all are uniquely 
# identified 

df.fueltype = df.fueltype.astype('category').cat.codes
df.aspiration = df.aspiration.astype('category').cat.codes
df.doornumber = df.doornumber.astype('category').cat.codes
df.carbody = df.carbody.astype('category').cat.codes
df.drivewheel = df.drivewheel.astype('category').cat.codes
df.enginelocation = df.enginelocation.astype('category').cat.codes
df.enginetype = df.enginetype.astype('category').cat.codes
df.cylindernumber = df.cylindernumber.astype('category').cat.codes
df.fuelsystem = df.fuelsystem.astype('category').cat.codes
```

```python
df = df.drop(['car_ID'], axis = 1)
```

```python
df.columns
```

```python
df.corr()
```

```python
# drop the column which are not used --
df.drop(['CarName','fueltype','citympg','curbweight'],axis=1,inplace=True)
```

```python
df.dtypes
```

# build model  ----

```python
X=df.drop(["price"],axis=1)
y=df["price"]
```

```python
X
```

# doing the train_test split --

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2
                                               ,random_state=101)
```

```python
model=LinearRegression() # build the linear regreesion model for predicting the price of car 
model.fit(x_train,y_train)
```

```python
# here let predicting the price of test data we have --
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
rmse
```

```python
ac=model.score(x_test,y_test)
```

```python
ac
```
