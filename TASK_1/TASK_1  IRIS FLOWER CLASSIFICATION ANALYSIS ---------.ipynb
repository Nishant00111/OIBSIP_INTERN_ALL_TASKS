# LOGISTIC REGRESSION --

```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = ('C:\ML  first data/iris_dataset.csv')
```

```python
df = pd.read_csv(data)
```

```python
df
```

```python
df['target'].unique()
```

## here three target is given so we drop one of them to implement logistic regression --

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# change the value of target into 0 and 1 as.. using mapping

df['target']=df['target'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
df.head()
```

```python
### Split dataset into independent and dependent features
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

## now, doing the train test split and import logistic regression -- 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
classifier
```

### NOW WE DOING HYPER-PARAMETER TUNNING FOR FINDING OUR PARAMETERS OF OUR LOGISTIC MODEL AS --

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,
                                  scoring='accuracy',cv=5)
```

```python
# TRAINING OUR DATA--
classifier_regressor.fit(X_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
# NOW DOING PREDICTION AND TEST DATA --

y_pred=classifier_regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test # actual values is here 
```

```python
## accuracy score

from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(classifier_regressor, X_test, y_test)
```
